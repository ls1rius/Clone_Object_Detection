```python
import pandas as pd
import string
import numpy as np
```

```python
df_dt = pd.read_csv('test_res.csv')
label_list = list(df_dt.columns[1:])
df_dt.columns=['image_id', 0, 1, 2]
df_dt['label_dt'] = df_dt[[0, 1, 2]].idxmax(1)
```

```python
df_gt = pd.read_csv('../data/test_day6.csv')
df_gt['label_gt'] = df_gt['label'].apply(lambda x: label_list.index(x.strip(string.digits)))
```
