```python
import cv2
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
import random
import time
import shutil
import numpy as np
import torchvision
```

```python
df_res = pd.read_csv('./train_stitch.csv')
```

```python
df_tmp = df_res.sample(1)
tmp_img_path = df_tmp['filename_gfp'].tolist()[0]
df_tmp = df_res[df_res['filename_gfp'] == tmp_img_path]
```

```python
df_tmp
```

```python
img = cv2.imread(os.path.join('./img_stitch_all', df_tmp['filename_gfp'].tolist()[0]))
```

```python
for i in range(len(df_tmp)):
    x1 = df_tmp.iloc[i]['x1']
    y1 = df_tmp.iloc[i]['y1']
    x2 = df_tmp.iloc[i]['x2']
    y2 = df_tmp.iloc[i]['y2']
    img = cv2.rectangle(img,
                        (int(x1-1), int(y1-1)),
                        (int(x2+1), int(y2+1)),
                        (255,0,0),
                        1)
```

```python
width, height = img.shape[1], img.shape[0]
img = np.ascontiguousarray(img)
fig = plt.figure(frameon=False)
canvas = fig.canvas
dpi = fig.get_dpi()
fig.set_size_inches((width + 1e-8) / dpi, (height + 1e-8) / dpi)
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax = plt.gca()
ax.axis('off')
plt.imshow(img)
```

```python
def get_detection_transforms():
#         transforms_list = []
    transforms_list = [
        torchvision.transforms.RandomHorizontalFlip(0.5),
        torchvision.transforms.RandomVerticalFlip(0.5),
        torchvision.transforms.RandomRotation(180),
        torchvision.transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0)
    ]
    return torchvision.transforms.Compose(transforms_list)
```

```python
detection_transforms = get_detection_transforms()
```

```python
img = torchvision.transforms.functional.to_tensor(img)
```

```python
for i in range(len(df_tmp)):
    x1 = df_tmp.iloc[i]['x1']
    y1 = df_tmp.iloc[i]['y1']
    x2 = df_tmp.iloc[i]['x2']
    y2 = df_tmp.iloc[i]['y2']
    obj_img = img[:, y1:y2, x1:x2]
    img[:, y1:y2, x1:x2] = detection_transforms(obj_img)
```

```python
img = img.permute(1,2,0)
```

```python
width, height = img.shape[1], img.shape[0]
img = np.ascontiguousarray(img)
fig = plt.figure(frameon=False)
canvas = fig.canvas
dpi = fig.get_dpi()
fig.set_size_inches((width + 1e-8) / dpi, (height + 1e-8) / dpi)
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
ax = plt.gca()
ax.axis('off')
plt.imshow(img)
```
